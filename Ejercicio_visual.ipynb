```python
#Relacion entre edad con la familia y los amigos.

Year = df["Year"]
relacion_padres = df['Time spent with with parents, siblings and other family, by age of respondent (United States)']
relacion_amigos = df['Time spent with friends, by age of respondent (United States)']

# Crear la figura y los subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 8))

# Gr√°fico de dispersi√≥n para la relaci√≥n con los padres
sns.scatterplot(x = Year, y=relacion_padres, color='red', ax=axes[0])
axes[0].set_xlabel('Edad')
axes[0].set_ylabel('Minutos por d√≠a')
axes[0].set_title('Relaci√≥n con los padres vs edad')
axes[0].set_xlim(0,90 + 5)
axes[0].set_ylim(0,250)



# Gr√°fico de dispersi√≥n para la relaci√≥n con los amigos
sns.scatterplot(x = Year, y=relacion_amigos, color='blue', ax=axes[1])
axes[1].set_xlabel('Edad')
axes[1].set_ylabel('Minutos por d√≠a')
axes[1].set_title('Relaci√≥n con los amigos vs edad')
axes[1].set_xlim(0,90 + 5)
axes[1].set_ylim(0,250)
```

En estos 2 gr√°ficos de dispersi√≥n podemos ver que las personas de los 35 en adelante pasan muy pocos minutos al d√≠a con sus amigos y sus padres.  
La relaci√≥n con los amigos no llega ni a los 50 min por d√≠a, y con los padres casi una hora. Se podr√≠a decir que entre 35 hasta morir las personas dedican el mismo tiempo a los amigos y sus padres

```python
#Realcion de edad con el trabajo y con la pareja 

Year = df["Year"]
relacion_trabajo = df["Time spent with coworkers, by age of respondent (United States)"]
relacion_pareja = df["Time spent with partner, by age of respondent (United States)"]
```

```python

# Crear la figura y los subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 8))

# Gr√°fico de dispersi√≥n para la relaci√≥n con los padres
sns.scatterplot(x = Year, y=relacion_trabajo, color='red', ax=axes[0])
axes[0].set_xlabel('Edad')
axes[0].set_ylabel('Minutos por d√≠a')
axes[0].set_title('Relaci√≥n con los compa√±eros de trabajo vs edad')
axes[0].set_xlim(0,90 + 5)
axes[0].set_ylim(0,250)



# Gr√°fico de dispersi√≥n para la relaci√≥n con los amigos
sns.scatterplot(x = Year, y=relacion_pareja, color='blue', ax=axes[1])
axes[1].set_xlabel('Edad')
axes[1].set_ylabel('Minutos por d√≠a')
axes[1].set_title('Relaci√≥n con la pareja vs edad')
axes[1].set_xlim(0,90 + 5)
axes[1].set_ylim(0,250)
```

En este grafico nos demuestra donde pasan mas tiempo las personas entre 25 a hasta los 80. Con sus compa√±eros de trabajo y su pareja. No podemos evaluar cuantos de los estudiados ten√≠an o no pareja o cuantos trabajan o no, pero si nos da una curva opuesta de los gr√°ficos anterior. 
De esto podemos deducir que las personas entre 30 a 70 a√±o de edad pasan m√°s tiempo con su pareja o trabajadores que con sus seres queridos (familias y amigos)
